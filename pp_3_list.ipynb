```python
# Ex no.1:

students = ["Arun", "Thiru", "Venkey", "Thiru", "Adhu"]
print("\nOriginal Student List:", students)

students.append("Suresh")
print("\nAfter adding a student:", students)

backup_students = students.copy()
print("\nBackup copy:", backup_students)

print("\nThiru appears:", students.count("Thiru"), "times")

transfer_students = ["Vishal"]
students.append(transfer_students)
print("\nAfter merging transfer students:", students)

print("\nPosition of Adhu:", students.index("Adhu"))

students.insert(2, "Prior_student")
print("\nAfter inserting :", students)

removed_student = students.pop()
print("\nRemoved student:", removed_student)

students.remove("Thiru")
print("\nAfter removing withdrawn student:", students)

print("\nReversed list:", students[::-1])

students.sort()
print("\nSorted list:", students)

students.clear()
print("\nAfter clearing records:", students)
```
