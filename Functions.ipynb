```python
def count_vowels(s):
    vowels = "aeiou"
    return sum(1 for ch in s.lower() if ch in vowels)

print(count_vowels("Education"))
```

```python
students = {
    "Anu": [78, 85, 90],
    "Bala": [60, 72, 68],
    "Charan": [88, 91, 84],
    "Divya": [45, 50, 40]
}

def calculate_average(marks):
    return sum(marks)/len(marks)

result = {}

for name, marks in students.items():
    avg = round(calculate_average(marks), 2)
    status = (lambda x: "Pass" if x >= 50 else "Fail")(avg)
    result[name] = {"Average": avg, "Result": status}

# Print passed students
print("Passed Students:")
for k,v in result.items():
    if v["Result"] == "Pass":
        print(k, v)

# Topper
topper = max(result, key=lambda x: result[x]["Average"])
print("Topper:", topper)
```

```python
def count_digits(n):
    if n == 0:
        return 0
    return 1 + count_digits(n//10)

def sum_digits(n):
    if n == 0:
        return 0
    return n%10 + sum_digits(n//10)

def max_digit(n):
    if n < 10:
        return n
    return max(n%10, max_digit(n//10))

def reverse_num(n, rev=0):
    if n == 0:
        return rev
    return reverse_num(n//10, rev*10 + n%10)

num = 5832

print("Digits:", count_digits(num))
print("Sum:", sum_digits(num))
print("Max Digit:", max_digit(num))
print("Reverse:", reverse_num(num))
```
