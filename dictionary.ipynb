```python
Dict = {}
print("Empty Dictionary: ")
print(Dict)

# Adding elements to dictionary one at a time
Dict[0] = 'Peter'
Dict[2] = 'Joseph'
Dict[3] = 'Ricky'
print("\nDictionary after adding 3 elements: ")
print(Dict)

# Adding set of values with a single Key
# The Emp_ages doesn't exist to dictionary

Dict['Emp_ages'] = 20, 33, 24
print("\nDictionary after adding 3 elements: ")
print(Dict)

# Updating existing Key's Value
Dict[3] = 'JavaTpoint'
print("\nUpdated key value: ")
print(Dict)
```

```python
books = {
    101: {"title": "Python", "author": "Rossum", "available_copies": 3},
    102: {"title": "AI", "author": "Russell", "available_copies": 5}
}

print("Original Books:", books)

# Backup
backup = books.copy()
print("\nBackup:", backup)

# Default categories
categories = dict.fromkeys(["Science", "Arts", "Engineering"], 0)
print("\nCategories:", categories)

# Safe retrieval
print("\nAvailable copies of book 101:", books.get(101, 0))
print("Available copies of book 999:", books.get(999, 0))

# Display all records
print("\nAll Book Details:")
for k, v in books.items():
    print(k, v)

# Only IDs
print("\nBook IDs:", list(books.keys()))

# Only available copies
copies = [b["available_copies"] for b in books.values()]
print("Available Copies:", copies)

# Remove discarded book
books.pop(101)
print("\nAfter removing book 101:", books)

# Remove last added book
books.popitem()
print("After popitem:", books)

# Ensure missing category
categories.setdefault("Medical", 0)
print("\nUpdated Categories:", categories)

# Merge shipment
new_books = {103: {"title": "ML", "author": "Tom", "available_copies": 4}}
books.update(new_books)
print("\nAfter Shipment:", books)

# Clear all records
books.clear()
print("\nAfter Clearing Books:", books)
```
